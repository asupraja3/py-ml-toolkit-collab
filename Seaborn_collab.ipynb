Seaborn is a Python data visualization library built on top of matplotlib that provides a high-level interface for creating attractive and informative statistical graphics.


üîç**Why Use Seaborn?**
* Simpler syntax than matplotlib for complex plots
* Built-in themes and color palettes
* Integrates well with pandas DataFrames
* Ideal for statistical data exploration  

‚úÖ When to Use Seaborn
* For quick exploratory data analysis
* When working with tabular data (pandas)
* For statistical plots like KDE, violin, box, and heatmaps

‚úÖ When to Use Matplotlib
* When you need precise control over every plot element (ticks, grids, fonts)
* For non-statistical or scientific plotting
* When creating complex or multi-subplot visualizations

The small vertical line you see on top of bars in a Seaborn bar plot is called an error bar.  
üìè What Are Error Bars?
* They represent the uncertainty or variation in the data.
* By default, Seaborn‚Äôs barplot() shows 95% confidence intervals (CI) around the mean.
* It visually communicates how precise or variable the estimate is.

```python
plt.figure(figsize=(9,3))
sns.barplot(x='class', y='fare', data=titanic_df, hue='sex')
```

**HeatMAp**  
A heatmap in Seaborn is a color-coded matrix that helps visualize the magnitude of values across a 2D grid ‚Äî commonly used to display correlation matrices, confusion matrices, or any tabular numerical data.  
**üî• What Is a Heatmap?**
* Each cell in the matrix is color-shaded based on its value.
* Darker or lighter colors indicate higher or lower values (depending on the color map).
* Very useful for spotting patterns, trends, outliers, or correlations visually.

**‚úÖ When to Use Heatmaps**  
* Visualizing correlation matrices
* Showing confusion matrices in classification
* Viewing pivot tables or aggregated values
* Detecting missing data patterns (e.g., with sns.heatmap(df.isnull()))

```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
flight_df = flight_df.pivot(index="month", columns="year", values="passengers")
flight_df.head()
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(flight_df)
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(flight_df, annot=True, fmt='d')
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(flight_df, annot=True, fmt='d',
              linecolor='k', linewidths='1')
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(flight_df, annot=True, fmt='d',
              linecolor='k', linewidths='1',
            cmap='YlGnBu')
```

```python
#added all code for the bar orientation change
plt.figure(figsize=(6,6))
grid_kws = {"height_ratios": (.4, .03), "hspace": 0.4}
f, (ax,cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws, figsize=(6,6))
sns.heatmap(flight_df, annot=True, fmt='d',
              linecolor='k', linewidths='1',
              cmap='YlGnBu', cbar_kws={'label': 'Passengers','orientation':'horizontal'},
              ax =ax , cbar_ax = cbar_ax)
```

In a Seaborn heatmap, when used with a correlation matrix, correlation refers to the statistical relationship between two numeric variables ‚Äî how closely they move together.  
**üìà What is Correlation?**
* Measures the strength and direction of a linear relationship between two variables.
* Calculated using Pearson correlation coefficient by default in pandas (.corr()).

```python
titanic = sns.load_dataset('titanic')
# Select only numerical columns for correlation
num_cols = titanic.select_dtypes(include=['number'])
corr = num_cols.corr()
```

```python
plt.figure(figsize=(5, 5))
sns.heatmap(corr, annot=True, center=0)
```

**fare vs pclass:**	Negative correlation (higher class ‚Üí lower fare)  
**age vs survived:**	Slight negative correlation (older = lower survival)  
**sibsp vs parch:** Weak positive correlation (family size effects)

# üîó CLUSTERMAP `clustermap()`
The `sns.clustermap()` function plots a matrix dataset as a **hierarchically-clustered heatmap** with dendrograms.

It‚Äôs commonly used for:
- **Correlation matrices**
- **Gene expression heatmaps**
- **Pattern detection in features**

**Seaborn‚Äôs clustermap() is a function that creates a heatmap with hierarchical clustering of both rows and columns.  **
It is a combination of:

üî• A heatmap showing values in a matrix (e.g., correlations, features, etc.)

üå≥ Dendrograms showing how rows/columns are grouped based on similarity


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create sample data (10 rows √ó 8 features)
np.random.seed(42)
data = np.random.rand(10, 8)

# Row and column labels
row_labels = [f"Sample_{i}" for i in range(10)]
col_labels = [f"Feature_{i}" for i in range(8)]

# Convert to DataFrame
import pandas as pd
df = pd.DataFrame(data, index=row_labels, columns=col_labels)
```

## üîç `sns.clustermap()` ‚Äì Hierarchical Cluster Heatmap
Creates a clustered heatmap with optional annotations and color scaling.


```python

plt.figure(figsize=(3,3))
sns.clustermap(df)
plt.show()
```

## ‚öôÔ∏è Key Parameters of `sns.clustermap()`

- `data`: (DataFrame or 2D array) Matrix to plot.
- `method`: (str) Linkage method for clustering. E.g., 'single', 'complete', 'average', 'ward'.
- `metric`: (str) Distance metric for clustering. E.g., 'euclidean', 'cityblock', 'cosine'.
- `z_score`: (int) Standardize rows (0) or columns (1) before clustering.
- `standard_scale`: (int) Normalize rows (0) or columns (1) to 0 mean and unit variance.
- `figsize`: (tuple) Size of entire figure.
- `row_cluster`: (bool) Whether to cluster rows.
- `col_cluster`: (bool) Whether to cluster columns.
- `row_colors`: (Series or array) Colors to label rows.
- `col_colors`: (Series or array) Colors to label columns.
- `cmap`: (str or colormap) Colormap for the heatmap (e.g., 'viridis', 'coolwarm').
- `linewidths`: (float) Width of lines between cells.
- `dendrogram_ratio`: (tuple) Ratios of dendrogram size to matrix size.
- `colors_ratio`: (tuple) Ratios for the row/column color bars.
- `cbar_pos`: (tuple) Position of color bar: (x, y, width, height).
